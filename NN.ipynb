```python
# Import relevant libraries
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np

# Separate target variable (size_class) and features (X)
y = lightning_df['size_class']
X = lightning_df.drop(columns=['size_class'])

# Handle missing values by imputing with the mean
X = X.fillna(X.mean())

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)

# Arrange variables in descending order of explained variance
sorted_indices = np.argsort(pca.explained_variance_)[::-1]
sorted_features = [X.columns[i] for i in sorted_indices]

# Plot cumulative variance explained
plt.figure(figsize=(10, 6))
plt.plot(
    range(1, len(cumulative_variance) + 1),
    cumulative_variance,
    marker='o', linestyle='--', color='b'
)
plt.title('Cumulative Variance Explained by Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Variance Explained')
plt.grid(True)
plt.show()
```

```python
# Get the PCA components
loadings = pd.DataFrame(
    pca.components_.T,
    index=X.columns,
    columns=[f'PC{i+1}' for i in range(pca.n_components_)]
)

# Calculate the overall contribution of each feature
# Use the absolute sum of loadings across all principal components
feature_importance = loadings.abs().sum(axis=1)

# Sort features by importance in descending order
top_features = feature_importance.sort_values(ascending=False).head(10)

# Display the top 10 features
print("Top 10 Features Contributing to PCA:")
print(top_features)

# Optional: Visualize the top features
top_features.plot(kind='barh', figsize=(10, 6), color='skyblue')
plt.title('Top 10 Features Contributing to PCA')
plt.xlabel('Cumulative Contribution to PCA')
plt.ylabel('Features')
plt.gca().invert_yaxis()  # Invert axis for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# prompt: create a neural network to predict size_class and generate a confusion matrix with %age, shuffle off, split the data from upto 31st dec 2015 for train and rest for validation

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'lightning_df' and other necessary variables are defined from the previous code

# Convert 'discovered_date' to datetime objects
lightning_df['discovered_date'] = pd.to_datetime(lightning_df['discovered_date'])

# Split data based on date
train_data = lightning_df[lightning_df['discovered_date'] <= '2015-12-31']
val_data = lightning_df[lightning_df['discovered_date'] > '2015-12-31']

# Separate features and target variable for both train and validation sets
X_train = train_data.drop(columns=['size_class', 'discovered_date'])
y_train = train_data['size_class']
X_val = val_data.drop(columns=['size_class', 'discovered_date'])
y_val = val_data['size_class']

# Handle missing values (if any) - you might need a more sophisticated strategy
X_train = X_train.fillna(X_train.mean())
X_val = X_val.fillna(X_val.mean())

# Scale the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)

# Initialize and train the MLP classifier
mlp = MLPClassifier(hidden_layer_sizes=(100,50), max_iter=500, random_state=42)  # Adjust hyperparameters as needed
mlp.fit(X_train_scaled, y_train)

# Make predictions on the validation set
y_pred = mlp.predict(X_val_scaled)

# Create and display the confusion matrix with percentages
cm = confusion_matrix(y_val, y_pred, normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y_val))
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix (Percentage)')
plt.show()
```

```python
# Ensure necessary imports
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Ensure the 'year' column exists
if 'year' not in lightning_df.columns:
    raise KeyError("'year' column is missing from the dataset")

# Prepare the data
X = lightning_df.drop('year', axis=1, errors='ignore')  # Features
y = lightning_df['year']  # Target variable

# Convert discovery_date if present
if 'discovery_date' in X.columns:
    X['discovery_date'] = pd.to_datetime(X['discovery_date'], errors='coerce')
    X['discovery_date_timestamp'] = X['discovery_date'].view('int64') // 10**9
    X.drop(columns=['discovery_date'], inplace=True)

# Split data into training (1996-2015) and testing (2016-2020)
X_train = X[lightning_df['year'] <= 2015]
X_test = X[lightning_df['year'] > 2015]
y_train = y[lightning_df['year'] <= 2015]
y_test = y[lightning_df['year'] > 2015]

# Identify non-numeric columns
non_numeric_columns = X_train.select_dtypes(include=['object']).columns
if len(non_numeric_columns) > 0:
    print(f"Converting non-numeric columns: {non_numeric_columns}")
    # Convert categorical columns to numeric
    X_train = pd.get_dummies(X_train, columns=non_numeric_columns, drop_first=True)
    X_test = pd.get_dummies(X_test, columns=non_numeric_columns, drop_first=True)
    # Reindex X_test to match X_train
    X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

import pandas as pd

# Assuming X_train is your feature set
# Check for datetime columns
datetime_columns = X_train.select_dtypes(include=['datetime64']).columns

# Convert datetime columns to numeric features (e.g., extract year, month, day)
for col in datetime_columns:
    X_train[col + '_year'] = X_train[col].dt.year
    X_train[col + '_month'] = X_train[col].dt.month
    X_train[col + '_day'] = X_train[col].dt.day
    X_train.drop(columns=[col], inplace=True)

# Repeat for X_test if applicable
datetime_columns_test = X_test.select_dtypes(include=['datetime64']).columns
for col in datetime_columns_test:
    X_test[col + '_year'] = X_test[col].dt.year
    X_test[col + '_month'] = X_test[col].dt.month
    X_test[col + '_day'] = X_test[col].dt.day
    X_test.drop(columns=[col], inplace=True)

# Train the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)

# Make predictions
y_pred = rf_classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')  # For multi-class classification
recall = recall_score(y_test, y_pred, average='weighted')  # For multi-class classification
f1 = f1_score(y_test, y_pred, average='weighted')  # For multi-class classification

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Print the results
print(f"Accuracy of the Random Forest model: {accuracy:.4f}")
print(f"Precision of the Random Forest model: {precision:.4f}")
print(f"Recall of the Random Forest model: {recall:.4f}")
print(f"F1-Score of the Random Forest model: {f1:.4f}")
print("\nConfusion Matrix:")
print(conf_matrix)
```
