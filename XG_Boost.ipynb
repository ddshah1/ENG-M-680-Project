```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score
import xgboost as xgb
from shapely import wkt

# Load the dataset
file_path = '/content/drive/Shareddrives/ENG M 680/processed_wildfire_df.csv'  # Replace with your actual file path
wildfire_df = pd.read_csv(file_path)

# Step 1: Create target variable - lightning caused fire
# Use `general_cause` column to determine if a fire was caused by lightning
if 'general_cause' in wildfire_df.columns:
    wildfire_df['is_lightning_caused'] = (wildfire_df['general_cause'].str.lower() == 'lightning').astype(int)
else:
    print("Error: The column 'general_cause' does not exist in the dataset.")
    wildfire_df['is_lightning_caused'] = 0  # Default to 0 if `general_cause` isn't available

# Verify the target variable distribution
print("Target variable distribution:\n", wildfire_df['is_lightning_caused'].value_counts())

# Split data into training and testing sets based on the year (<2015 for training, >=2015 for testing)
train_data = wildfire_df[wildfire_df['year'] < 2015]
test_data = wildfire_df[wildfire_df['year'] >= 2015]

# Select relevant features (excluding cause-related columns like `general_cause`, `detection_agent_type`)
features = ['current_size', 'size_class', 'FA_NAME', 'smoke_type', 'smoke_color',
            'smoke_base_size', 'ia_start_size_obj', 'hour_of_day']
target = 'is_lightning_caused'

# Handle missing values by filling with a placeholder
train_data = train_data[features + [target]].fillna('Unknown')
test_data = test_data[features + [target]].fillna('Unknown')

# Encode categorical features
le = LabelEncoder()
for col in train_data.select_dtypes(include=['object']).columns:
    train_data[col] = le.fit_transform(train_data[col])
    test_data[col] = le.transform(test_data[col])

# Separate features and target
X_train, y_train = train_data[features], train_data[target]
X_test, y_test = test_data[features], test_data[target]

# Step 2: Train the XGBoost classifier with limited rounds for efficiency
xgb_classifier = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', max_depth=3, n_estimators=100)
xgb_classifier.fit(X_train, y_train)

# Predict on the test set and evaluate
y_pred = xgb_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, output_dict=True)

# Display accuracy and detailed classification report metrics
print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(classification_report(y_test, y_pred))



```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Non-Lightning", "Lightning"], yticklabels=["Non-Lightning", "Lightning"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix of Lightning-Caused Fire Prediction")
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import xgboost as xgb
import seaborn as sns
import matplotlib.pyplot as plt
from shapely import wkt

# Load the dataset
file_path = '/content/drive/Shareddrives/ENG M 680/processed_wildfire_df.csv'  # Replace with your actual file path
wildfire_df = pd.read_csv(file_path)

# Step 1: Create target variable - lightning caused fire
# Use `general_cause` column to determine if a fire was caused by lightning
if 'general_cause' in wildfire_df.columns:
    wildfire_df['is_lightning_caused'] = (wildfire_df['general_cause'].str.lower() == 'lightning').astype(int)
else:
    print("Error: The column 'general_cause' does not exist in the dataset.")
    wildfire_df['is_lightning_caused'] = 0  # Default to 0 if `general_cause` isn't available

# Verify the target variable distribution
print("Target variable distribution:\n", wildfire_df['is_lightning_caused'].value_counts())

# Split data into training and testing sets based on the year (<2015 for training, >=2015 for testing)
train_data = wildfire_df[wildfire_df['year'] < 2015]
test_data = wildfire_df[wildfire_df['year'] >= 2015]

# Select relevant features (excluding cause-related columns like `general_cause`, `detection_agent_type`)
features = ['current_size', 'size_class', 'FA_NAME', 'smoke_type', 'smoke_color',
            'smoke_base_size', 'ia_start_size_obj', 'hour_of_day']
target = 'is_lightning_caused'

# Handle missing values by filling with a placeholder
train_data = train_data[features + [target]].fillna('Unknown')
test_data = test_data[features + [target]].fillna('Unknown')

# Encode categorical features
le = LabelEncoder()
for col in train_data.select_dtypes(include=['object']).columns:
    train_data[col] = le.fit_transform(train_data[col])
    test_data[col] = le.transform(test_data[col])

# Separate features and target
X_train, y_train = train_data[features], train_data[target]
X_test, y_test = test_data[features], test_data[target]

# Step 2: Hyperparameter Tuning with GridSearchCV
param_grid = {
    'max_depth': [3, 5, 7],            # Controls tree depth
    'learning_rate': [0.01, 0.1, 0.2], # Step size shrinkage
    'n_estimators': [100, 200, 300],    # Number of boosting rounds
    'subsample': [0.7, 0.8, 1.0],      # Subsampling of training samples
    'colsample_bytree': [0.7, 0.8, 1.0], # Fraction of features for each tree
    'gamma': [0, 0.1, 0.2]             # Minimum loss reduction to make further partition
}

# Initialize the XGBoost classifier
xgb_classifier = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')

# Perform grid search
grid_search = GridSearchCV(estimator=xgb_classifier, param_grid=param_grid, scoring='accuracy', cv=3, n_jobs=-1, verbose=1)
grid_search.fit(X_train, y_train)

# Print best parameters and best score
print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)

# Train XGBoost with best parameters
best_xgb_classifier = grid_search.best_estimator_
best_xgb_classifier.fit(X_train, y_train)

# Predict on the test set and evaluate
y_pred = best_xgb_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

# Display accuracy and detailed classification report metrics
print(f"Test Set Accuracy: {accuracy}")
print("Classification Report:")
print(report)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Non-Lightning", "Lightning"], yticklabels=["Non-Lightning", "Lightning"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix of Lightning-Caused Fire Prediction")
plt.show()
```
